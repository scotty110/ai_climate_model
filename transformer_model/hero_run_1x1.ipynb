## Transformer Based Model

```python
class InnerTransformer(nn.Module):
    def __init__(self, input_dim:int, output_dim:int, num_layers:int=2):
        super(InnerTransformer, self).__init__()
        '''
        Args:
            - input_dim (int): Number of input features
            - output_dim (int): Number of output features
            - num_layers (int): Number of transformer layers, default is 1
        '''
        self.input_dim = input_dim
        self.output_dim = output_dim
        self.num_layers = num_layers

        # Build transformer layers
        self.layers = nn.ModuleList([nn.TransformerEncoderLayer(d_model=self.input_dim, nhead=1) for _ in range(self.num_layers)]) # Make it simple for now

        # Linear layer to output
        self.linear = nn.Linear(self.input_dim, self.output_dim)

    def forward(self, x:torch.Tensor) -> torch.Tensor:
        for layer in self.layers:
            x = layer(x)
        return x
```

```python
class FullTrans(nn.Module):
    def __init__(self, input_shape:list[int]=[32,3], inner_shape:int=int(32*4), out_dim:int=int(2), num_layers:int=4):
        super(FullTrans, self).__init__()
        '''
        Build a DepthWiseConv1d model followed by a Transformer model followed by an output layer(tbd)
        Args:
            - input_shape (list): Shape of the input tensor, default is [32,3] (32 cells, 3 features)
            - inner_shape (int): Number of features to pass to the transformer, default is 32*2
            - out_dim (int): Number of output features, default is 32*2 (32 cells, 2 features)
            - num_layers (int): Number of transformer layers, default is 2
        '''
        self.encoding = nn.Linear(input_shape[1], inner_shape) 
        self.transformer = InnerTransformer(input_dim=inner_shape, output_dim=inner_shape, num_layers=num_layers)
        self.output_cov = nn.Conv1d(in_channels=inner_shape, out_channels=32, kernel_size=1)
        self.output_linear = nn.Linear(32, out_dim)

    def forward(self, x:torch.Tensor) -> torch.Tensor:
        x = self.encoding(x)
        x = self.transformer(x)
        x = x.permute(0, 2, 1)
        x = self.output_cov(x)
        x = x.squeeze()
        x = self.output_linear(x)
        x = x.view(-1, 32, 2) # Maybe make args??? IDK
        return x
```

Method to load hdf5 file of the processed weather data

```python
def load_hdf5(filename:str):
    '''
    Load data from an HDF5 file and return a list of dictionaries.
    Inputs:
        - filename (str): Path to the HDF5 file.
    Outputs:
        - data (list): List of dictionaries, where each dictionary represents an entry in the original list.
    '''
    data = []  # List to hold dictionaries
    with h5.File(filename, 'r') as f:
        # Iterate through groups (each representing an entry in the original list)
        for group_name in f:
            group = f[group_name]
            # Reconstruct dictionary from datasets and attributes
            entry = {
                # Attributes (metadata)
                'day': group.attrs['day'],
                'region': group.attrs['region'],
                'time': group.attrs['time'],

                # groups (numpy arrays)
                'landmass': group['landmass'][...],  # Use [...] to read the full dataset
                'x': group['x'][...],
                'y': group['y'][...],
            }
            data.append(entry)
    return data
```

Generate stacks to train

```python
def stack_data(data:list[dict], key:str) -> torch.Tensor:
    return torch.stack([torch.tensor(entry[key]) for entry in data])


def generate_stacks(data:list[dict]) -> tuple[torch.Tensor, torch.Tensor, torch.Tensor]:
    '''
    Create a PyTorch DataLoader from the data.
    Inputs:
        - data (list): List of dictionaries, where each dictionary represents an entry in the original list.
    Outputs:
        - x (torch.Tensor): Tensor of x-coordinate data.
        - y (torch.Tensor): Tensor of y-coordinate data.
    '''
    x = stack_data(data, 'x')
    x = x.transpose(2, 1)
    x = x.view(-1, 32, 3, 4)

    # Take the mean across the last dimension (think cam is 16 columns not 2x2)
    x = x.mean(dim=-1)
    #x = x.reshape(-1, 32, 3)

    y = stack_data(data, 'y')
    y = y.transpose(2, 1)
    y = y.view(-1, 32, 2, 4)
    
    # Take the mean across the last dimension (think cam is 16 columns not 2x2)
    y = y.mean(dim=-1)
    #y = y.reshape(-1, 32, 2)
    
    return (x, y)


class weather_dataset(Dataset):
    '''
    PyTorch Dataset class for weather data.
    '''
    def __init__(self, data:list[dict]):
        self.x, self.y = generate_stacks(data)
        self.length = len(self.x)

    def __len__(self):
        return self.length

    def __getitem__(self, idx):
        return (self.x[idx], self.y[idx])
```

Partition training data into train and validation sets.

```python
def get_dataloaders(fname:str, batch_size:int, split:int) -> tuple[DataLoader, DataLoader]:
    '''
    Create PyTorch DataLoader objects for training and validation data.
    Inputs:
        - fname (str): Path to the HDF5 file.
        - batch_size (int): Batch size for the DataLoader objects.
        - split (float): Fraction of the data to use for training.     
    Outputs:
        - train_loader (torch.utils.data.DataLoader): DataLoader for training data.
        - test_loader (torch.utils.data.DataLoader): DataLoader for test data.
    '''
    # Load data and create tensor 
    data = load_hdf5(fname)
    dataset = weather_dataset(data)
    
    train_size = int(split * len(dataset))
    val_size = len(dataset) - train_size
    
    # Split data into training and validation sets
    train_dataset, test_dataset = torch.utils.data.random_split(dataset, [train_size, val_size])

    # Create DataLoader objects
    train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True, num_workers=8, pin_memory=True)
    test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=True, num_workers=8, pin_memory=True)
    return train_loader, test_loader
```

## Model Block

Want to have normalization layer (x) -> reshape -> NN -> reshape - > denorm layer (y). 

```python
class NormalizationLayer(nn.Module):
    def __init__(self, data: torch.Tensor):
        super(NormalizationLayer, self).__init__()

        # Compute mean and std along the batch_size, x1, and x2 dimensions
        self.mean = data.mean(dim=(0, 1), keepdim=True).to(device, dtype=dtype) 
        self.std = data.std(dim=(0, 1), keepdim=True).to(device, dtype=dtype) 

    def forward(self, x:torch.Tensor) -> torch.Tensor:
        # Normalize the data
        normalized_data = (x - self.mean) / (self.std + 1e-9)  # Add a small constant to avoid division by zero
        return normalized_data
```

```python
class DenormalizationLayer(nn.Module):
    def __init__(self, data: torch.Tensor):
        super(DenormalizationLayer, self).__init__()

        # Compute mean and std along the batch_size, x1, and x2 dimensions
        self.mean = data.mean(dim=(0, 1), keepdim=True).to(device, dtype=dtype)  
        self.std = data.std(dim=(0, 1), keepdim=True).to(device, dtype=dtype) 

    def forward(self, x:torch.Tensor) -> torch.Tensor:
        # Denormalize the data
        denormalized_data = x * (self.std + 1e-9) + self.mean
        return denormalized_data
```

```python
class CompleteModel(nn.Module):
    def __init__(self, norm_layer:NormalizationLayer, neural_network:FullTrans, denorm_layer:DenormalizationLayer):
        super(CompleteModel, self).__init__()
        self.norm_layer = norm_layer
        self.neural_network = neural_network
        self.denorm_layer = denorm_layer

    def forward(self, x:torch.Tensor) -> torch.Tensor:
        x = self.norm_layer(x)

        x = self.neural_network(x)

        x = self.denorm_layer(x)
        return x
```

Only Norm/Denorm x and y

```python
def make_model(loader) -> CompleteModel:
    norm_layer = NormalizationLayer(loader.dataset.dataset.x)
    denorm_layer = DenormalizationLayer(loader.dataset.dataset.y)

    model = FullTrans().to(device, dtype=dtype) 

    network = CompleteModel(norm_layer, model, denorm_layer).to(device, dtype=dtype) 
    return network
```

## Training Loop 

Define training loop, use mixed percision training

```python
def train(model:nn.Module, dl:torch.utils.data.DataLoader, optim:torch.optim, loss:nn.Module) -> float:
    model.train()
    total_loss = .0
    scaler = GradScaler()

    for _, (x, y) in enumerate(dl):
        x = x.to(device, dtype=dtype) 
        y = y.to(device, dtype=dtype) 

        optim.zero_grad()
        with autocast(device_type='cuda', dtype=torch.float16):
            y_pred = model(x)
            l = loss(y_pred, y)
            total_loss += l.item()

        # Preform backpass
        scaler.scale(l).backward()
        scaler.step(optim)
        scaler.update()
    
    return total_loss / len(dl)
```

Eval Loop

```python
def eval(model:nn.Module, dl:torch.utils.data.DataLoader, loss:nn.Module) -> float:
    model.eval()
    total_loss = .0

    for _, (x, y) in enumerate(dl):
        x = x.to(device, dtype=dtype) 
        y = y.to(device, dtype=dtype)

        # Forward pass
        with autocast(device_type='cuda', dtype=torch.float16):
            y_pred = model(x)
            l = loss(y_pred, y)
            total_loss += l.item()

    return total_loss / len(dl)
```

## Training

```python
folds = 3
models = {}

test_ds = weather_dataset(load_hdf5(test_name)) 
test_loader = DataLoader(test_ds, batch_size=1024, shuffle=True, num_workers=8, pin_memory=True)

for i in range(folds):
    train_loader, val_loader = get_dataloaders(train_name, 1024, 0.6)

    loss_fn = nn.MSELoss()
    model = make_model(train_loader) 
    optimizer = optim.Adam(model.parameters(), lr=0.001)

    # Early Stopping 
    eval_loss = -1*float('inf') 
    train_loss  = float('inf') 
    j = 0
    print(f'Fold {i+1}')
    while train_loss > eval_loss:
        prev_eval_loss = eval_loss
        train_loss = train(model, train_loader, optimizer, loss_fn)
        eval_loss = eval(model, val_loader, loss_fn)
        print(f'\tEpoch {j} - Eval Loss: {eval_loss}')
        
        if abs(eval_loss - prev_eval_loss) < 0.005:
            print(f'\tEarly stopping at epoch {j} due to minimal change in eval loss')
            break
        j += 1

    # Test Loss
    test_loss = eval(model, test_loader, loss_fn)
    print(f'Fold {i+1} - Test Loss: {test_loss}')

    # Save Model
    model.eval()
    models[i] = model
```

```python
class cam_model_k(nn.Module):
    def __init__(self, model:nn.Module):
        super(cam_model_k, self).__init__()
        self.model = model

    def convert_to_cam(self, x:torch.Tensor) -> torch.Tensor:
        return

    def forward(self, t:torch.Tensor, p: torch.Tensor, m: torch.Tensor) -> tuple[torch.Tensor, torch.Tensor]:
        '''
        Args:
            - t (torch.Tensor): Tensor of temperature data, shape(Batch_Size, 16, 32)
            - p (torch.Tensor): Tensor of pressure, shape(Batch_Size, 16, 32)
            - m (torch.Tensor): Tensor of mixing ratios(?), shape(Batch_Size, 16, 32, 33)
        '''
        if torch.rand(1) < 0.5:
            m_new = m[:,:,:,0].squeeze(-1) # shape: (Batch_Size, 16, 32)

            #print(f'm_new shape: {m_new.shape}')
            con = torch.cat((t, m_new, p), dim=-1) # shape: (Batch_Size, 16, 32, 3)
            con = con.view(-1, 32, 3) # I think the 16 are single colums of cam

            # Do the model
            std = self.model(con)

            # Floor of 0
            std = torch.clamp(std, min=0)

            # Create a normal distribution
            mean = con[:, :, 0:2] 

            # Sample from the distribution
            noise = torch.randn_like(std[:, :, 0]).unsqueeze(-1).expand_as(std)  # Expand dim, temp and qv correlated.
            sample = mean + .01 * noise * std

            sample = sample.view(-1, 16, 32, 2)

            t_sample = sample[:,:,:,0].view(-1,16,32)
            qv_sample = sample[:,:,:,1].view(-1,16,32)
            m[:,:,:,0] = qv_sample
        
            m_diff = torch.abs( m[:,:,:,0] - qv_sample ).mean()
            print(f'Mixing Ratio Diff: {m_diff}')

        else:
            t_sample = t

        # Calc Difference between sample and real
        t_diff = torch.abs( t_sample - t ).mean()
        print(f'Temp Diff: {t_diff}')

        return (t_sample, m) 
```

```python
class cam_model(nn.Module):
    def __init__(self, model: nn.Module):
        super(cam_model, self).__init__()
        self.model = model

    def pressure_to_altitude(self, pressure_hPa: torch.Tensor) -> torch.Tensor:
        """
        Convert pressure (in hPa) to altitude (in meters) using the barometric formula.
        """
        # Constants for standard atmosphere (ICAO)
        p0 = 1013.25   # Sea-level standard atmospheric pressure, in hPa
        T0 = 288.15    # Standard temperature at sea level, in Kelvin
        L = 0.0065     # Temperature lapse rate, in K/m
        R = 8.31447    # Universal gas constant, in J/(mol*K)
        M = 0.0289644  # Molar mass of Earth's air, in kg/mol
        g = 9.80665    # Standard gravity, in m/s^2

        # Barometric formula to calculate altitude
        altitude_m = (T0 / L) * (1 - (pressure_hPa / p0) ** ((R * L) / (g * M)))
        return altitude_m

    def dse_to_kelvin(self, dse: torch.Tensor, geopotential: torch.Tensor) -> torch.Tensor:
        """
        Convert Dry Static Energy (DSE) to temperature in Kelvin.
        DSE = cp * T + g * z
        """
        cp = 1004.0    # Specific heat capacity of air at constant pressure, in J/(kg*K)
        g = 9.80665    # Gravitational acceleration, in m/s

        # Calculate temperature in Kelvin from DSE
        temperature_kelvin = (dse - g * geopotential) / cp
        return temperature_kelvin

    def kelvin_to_dse(self, t_kelvin: torch.Tensor, geopotential: torch.Tensor) -> torch.Tensor:
        """
        Convert temperature in Kelvin back to Dry Static Energy (DSE).
        DSE = cp * T + g * z
        """
        cp = 1004.0    # Specific heat capacity of air at constant pressure, in J/(kg*K)
        g = 9.80665    # Gravitational acceleration, in m/s

        # Calculate DSE from temperature in Kelvin
        dse = cp * t_kelvin + g * geopotential
        return dse

    def forward(self, t: torch.Tensor, p: torch.Tensor, m: torch.Tensor) -> tuple[torch.Tensor, torch.Tensor]:
        '''
        Args:
            - t (torch.Tensor): Tensor of DSE data, shape (Batch_Size, 16, 32)
            - p (torch.Tensor): Tensor of pressure data, shape (Batch_Size, 16, 32)
            - m (torch.Tensor): Tensor of mixing ratios, shape (Batch_Size, 16, 32, 33)

        Returns:
            - Tuple of (t_sample_dse, updated mixing ratios)
        '''
        # Pressure levels in hPa for the 32 vertical levels
        pressures_hPa = torch.tensor([
            3.643466, 7.59482, 14.356632, 24.61222, 35.92325, 43.19375,
            51.677499, 61.520498, 73.750958, 87.82123, 103.317127,
            121.547241, 142.994039, 168.22508, 197.908087, 232.828619,
            273.910817, 322.241902, 379.100904, 445.992574, 524.687175,
            609.778695, 691.38943, 763.404481, 820.858369, 859.534767,
            887.020249, 912.644547, 936.198398, 957.48548, 976.325407,
            992.556095
        ], device=t.device)  # Ensure it uses the same device as the input tensors

        # Convert pressure levels to altitude
        altitude_m = self.pressure_to_altitude(pressures_hPa)  # Shape: (32,)

        # Expand altitude_m to match t's shape
        altitude_m = altitude_m.view(1, 1, -1).expand(t.size(0), t.size(1), -1)  # Shape: (Batch_Size, 16, 32)

        # Convert DSE to temperature in Kelvin
        t_kelvin = self.dse_to_kelvin(t, altitude_m)  # Shape: (Batch_Size, 16, 32)

        # Extract the mixing ratio of water vapor (assume index 0)
        qv = m[:, :, :, 0]  # Shape: (Batch_Size, 16, 32)

        # Stack t_kelvin, qv, and p along a new dimension
        con = torch.stack((t_kelvin, qv, p), dim=-1)  # Shape: (Batch_Size, 16, 32, 3)

        # Flatten the batch and spatial dimensions to pass through the model
        con_flat = con.view(-1, 32, 3)  # Shape: (Batch_Size * 16, 32, 3)

        # Pass through the model
        std = self.model(con_flat)  # Model should output standard deviations, shape: (Batch_Size * 16, 32, 2)

        # Clamp std to be non-negative
        std = torch.clamp(std, min=0)

        # Assume mean is the t_kelvin and qv
        mean = con_flat[:, :, :2]  # Shape: (Batch_Size * 16, 32, 2)

        # Sample from the distribution
        noise = torch.randn_like(std)
        sample = mean + 0.01 * noise * std  # Shape: (Batch_Size * 16, 32, 2)

        # Reshape sample back to original dimensions
        sample = sample.view(t.size(0), t.size(1), 32, 2)  # Shape: (Batch_Size, 16, 32, 2)

        t_sample = sample[:, :, :, 0]  # Shape: (Batch_Size, 16, 32)
        qv_sample = sample[:, :, :, 1]  # Shape: (Batch_Size, 16, 32)

        # Update mixing ratio in m
        m[:, :, :, 0] = qv_sample

        # Convert t_sample (temperature in Kelvin) back to DSE
        t_sample_dse = self.kelvin_to_dse(t_sample, altitude_m)

        # Compute differences for logging
        m_diff = torch.abs(qv - qv_sample).mean()
        t_diff = torch.abs(t - t_sample_dse).mean()  # Compare original DSE with sampled DSE

        print(f'Mixing Ratio Diff: {m_diff.item()}')
        print(f'DSE Diff: {t_diff.item()}')

        return t_sample_dse, m
```
