# Can We Build an nn.Module for sampling 70xx4

Need to sample across 70 cells with given mean and std. 

```python
import torch
import torch.nn as nn
from torch.distributions import Normal
```

```python
class sample_distro(nn.Module):
    def __init__(self):
        super(sample_distro, self).__init__()
        
    def forward(self, mu, sigma):
        dist = Normal(mu, sigma) 
        s = dist.sample()

        # Do quick correction
        s[s < 0] = 0 
        return s
```

```python
mu = torch.rand(1, 70, 2, 4) #torch.tensor([0.0, 1.])
sigma = torch.rand(1, 70, 2, 4) #torch.tensor([0.1, 0.001])
```

```python
dist = sample_distro()
sample = dist(mu, sigma)
```
